```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
for i in range(200, 4500, 200):
    model_file = 'testModel.pt' + str(i)
    
    checkpoint = torch.load(model_file)
    batch_losses = checkpoint['batch_losses']
    mean_3bv_solves = checkpoint['3bv_solves']
    mean_3bv_solves = np.array(mean_3bv_solves) * 100
    
    plt.plot(batch_losses)
    plt.xlabel('batch index')
    plt.ylabel('batch loss')
    plt.show()
    
    plt.plot(mean_3bv_solves)
    plt.xlabel('batch index')
    plt.ylabel('3BV solved percent')
    plt.show()
```
